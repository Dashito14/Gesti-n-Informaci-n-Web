#### Ejercicio 1[3 puntos]
Se quiere simular una agenda de contactos de forma que de cada contacto se desea almacenar: nombre, apellidos, teléfono, y cuenta de email. 
El programa debe permitir:
 1) Añadir un contacto
 2) Eliminar un contacto.
 3) Modificar un contacto.
 4) Consultar datos de un contacto.
 5) Guardar agenda en un fichero.
 6) Leer agenda de un fichero
 7) Finalizar. Cuando se finaliza,automaticamente guarda la agenda en un fichero de texto.

```python
def aniadeContacto(nombre=0, apellidos=0, telefono=0, mail=0):
    if nombre == 0:
        print("-------------Añadir contacto-------------")
        
        print("Nombre: ")
        nombre= input ()
        while(nombre.isalpha() == False):
            print("Introduce un nombre real. Nombre:")
            nombre = input ()
            
        print("Apellidos: ")
        apellidos= input ()
        
        print("Telefono: ")
        telefono= input ()
        while(telefono.isdigit() == False):
            print("Introduce un valor numérico. Telefono:")
            telefono = input ()
            
        print("Mail: ")
        mail= input ()
        
    if nombre+apellidos in agenda:
        print("Ya hay un contacto con ese nombre")
    else:
        contacto= dict()
        contacto["nombre"]= nombre
        contacto["apellidos"]= apellidos
        contacto["telefono"]= telefono
        contacto["mail"]= mail
        agenda[nombre+apellidos]= contacto
        print("Contacto guardado correctamente")
        print("-------------------------------")
    
    
def eliminaContacto():
    print("-------------Eliminar contacto------------")
    print("Contacto a eliminar: ")
    print("Nombre: ")
    nombre= input ()
    print("Apellidos ")
    apellidos= input()
    if nombre+apellidos in agenda:
        del agenda[nombre+apellidos]
        print("El contacto se ha eliminado")
        print("---------------------------")
    else:
        print("No existe este contacto")
        print("-----------------------")
        
def modificarContacto():
    print("-------------Modificar contacto-------------")
    print("Contacto a modificar: ")
    print("Nombre: ")
    nombre= input ()
    print("Apellidos ")
    apellidos= input()
    if nombre+apellidos in agenda:
        print("¿Que desea cambiar(nombre/apellidos/telefono/mail)")
        op=input()
        if op=="nombre":
            print("Ingrese nuevo nombre")
            nuevonombre=input()
            contacto= dict()
            contacto["nombre"]= nuevonombre
            contacto["apellidos"]= apellidos
            contacto["telefono"]= agenda[nombre+apellidos]["telefono"]
            contacto["mail"]= agenda[nombre+apellidos]["mail"]
            agenda[nuevonombre+apellidos]= contacto
            del agenda[nombre+apellidos]
            print("Contacto modificado")
            
        elif op=="apellidos":
            print("Ingrese nuevos apellidos")
            nuevosapellidos=input()
            contacto= dict()
            contacto["nombre"]= nombre
            contacto["apellidos"]= nuevosapellidos
            contacto["telefono"]= agenda[nombre+apellidos]["telefono"]
            contacto["mail"]= agenda[nombre+apellidos]["mail"]
            agenda[nombre+nuevosapellidos]= contacto
            del agenda[nombre+apellidos]
            print("Contacto modificado")
            
        elif op=="telefono":
            print("Ingrese nuevo telefono")
            nuevotelefono=input()
            agenda[nombre+apellidos]["telefono"]=nuevotelefono
            print("Contacto modificado")
        
        elif op=="mail":
            print("Ingrese nuevo mail")
            nuevomail=input()
            agenda[nombre+apellidos]["mail"]=nuevomail
            print("Contacto modificado")
            
        else : 
            print("No existe la opcion escrita")
        
        
        print("-------------------")
    else:
        print("No existe este contacto")
        print("-----------------------")
        
def consultaContacto():
    print("-------------Consultar contacto-------------")
    print("Contacto a mostrar: ")
    print("Nombre: ")
    nombre= input ()
    print("Apellidos ")
    apellidos= input()
    contact=nombre+apellidos
    if contact in agenda:
        print("Datos de contacto:")
        print("Nombre: ", agenda[contact]["nombre"])
        print("Apellidos: ", agenda[contact]["apellidos"])
        print("Telefono: ", agenda[contact]["telefono"])
        print("Mail: ", agenda[contact]["mail"])
        print("--------------------------------")
    else:
        print("No existe ningun contacto con ese nombre")
        print("----------------------------------------")
    
def guardarAgenda():
    arc_write = open('agenda.txt', 'w')
    for contacto in agenda:
        for elemento in agenda[contacto]:
            arc_write.write(agenda[contacto][elemento]+" ")
        arc_write.write("\n")
    arc_write.close()
    
def cargarAgenda():    
    fichero= open("agenda.txt")
    for linea in fichero:
        contacto = linea.split()
        aniadeContacto(contacto[0],contacto[1],contacto[2],contacto[3])
    print("Agenda cargada correctamente")
    print("----------------------------")

def menu():
    opcion = 0
    while(opcion<1 or opcion>7):
        print("Seleccione una opción:")
        print("1. Añadir un contacto.")
        print("2. Eliminar un contacto.")
        print("3. Modificar un contacto.")
        print("4. Consultar un contacto.")
        print("5. Guardar agenda en un fichero.")
        print("6. Cargar agenda de un fichero.")
        print("7. Finalizar.")
        opcion = input()
        
        #Si introducimos un número
        if(opcion.isdigit()):
            opcion = int(opcion)
        
            #Si no seleccionamos un valor válido
            if(opcion < 1 or opcion > 7):
                print("Opción inválida, selecciona un valor entre 1 y 7.")
                print("-------------------------------------------------")
        
        #Si introducimos algo que no sea un número
        else:
            print("Introduce un número.")
            print("--------------------")
            #Volvemos a poner opcion a 0 para que entre en el bucle
            opcion = 0
            
    return opcion



opciones={
    1: aniadeContacto,
    2: eliminaContacto,
    3: modificarContacto,
    4: consultaContacto,
    5: guardarAgenda,
    6: cargarAgenda,
}



agenda = dict()
opcion = menu()

while(opcion >=1 and opcion <7):
    if opcion == 6:
        del agenda
        agenda = dict()
    opciones[opcion]()
    opcion = menu()
    
guardarAgenda()
print("Ha finalizado el programa y se ha guardado la agenda.")
```

Se pide implementar un programa que dado un sistema de 3 ecuaciones expresado en forma de una lista de listas donde cada lista representa una ecuación del sistema, devuelva como resultado los valores de las incognitas. 

Se espera que el sistema que se introduzca tenga solución, pero no hace falta realizar esta comprobación en el código del programa.

No se pueden usar ninguna función o método que calcule directamente el determinante.

```python
# Ejercicio 2
#
# Entrada de ejemplo
i =[[1,1,1,30],[1,3,-2,20],[1,1,-2, 0]]

def eliminarX(i):
    #Si el primer numero es negativo lo pasamos a positivo
    if(i[0][0] < 0):
        i[0][0] = i[0][0]*-1
        i[0][1] = i[0][1]*-1
        i[0][2] = i[0][2]*-1
        i[0][3] = i[0][3]*-1
       
    if(i[1][0] < 0):
        i[1][0] = i[1][0]*-1
        i[1][1] = i[1][1]*-1
        i[1][2] = i[1][2]*-1
        i[1][3] = i[1][3]*-1
   
    if(i[1][0] > 0):
        aux = i[0][0]
        #Cambiamos la primera funcion
        i[0][0] = i[0][0]*i[1][0]
        i[0][1] = i[0][1]*i[1][0]
        i[0][2] = i[0][2]*i[1][0]
        i[0][3] = i[0][3]*i[1][0]

        #Cambiamos la segunda funcion
        i[1][0] = i[1][0]*aux
        i[1][1] = i[1][1]*aux
        i[1][2] = i[1][2]*aux
        i[1][3] = i[1][3]*aux

        #Restamos para quitar la x en la segunda funcion
        i[1][0] = i[0][0] - i[1][0]
        i[1][1] = i[0][1] - i[1][1]
        i[1][2] = i[0][2] - i[1][2]
        i[1][3] = i[0][3] - i[1][3]

    #Si el primer numero es negativo lo pasamos a positivo
    if(i[2][0] < 0):
        i[2][0] = i[2][0]*-1
        i[2][1] = i[2][1]*-1
        i[2][2] = i[2][2]*-1
        i[2][3] = i[2][3]*-1
   
    if(i[2][0] > 0):
        aux = i[0][0]
        #Cambiamos la primera funcion
        i[0][0] = i[0][0]*i[2][0]
        i[0][1] = i[0][1]*i[2][0]
        i[0][2] = i[0][2]*i[2][0]
        i[0][3] = i[0][3]*i[2][0]

        #Cambiamos la segunda funcion
        i[2][0] = i[2][0]*aux
        i[2][1] = i[2][1]*aux
        i[2][2] = i[2][2]*aux
        i[2][3] = i[2][3]*aux

        #Restamos para quitar la x en la segunda funcion
        i[2][0] = i[0][0] - i[2][0]
        i[2][1] = i[0][1] - i[2][1]
        i[2][2] = i[0][2] - i[2][2]
        i[2][3] = i[0][3] - i[2][3]
       
   
    return i

def eliminarY(i):
    if(i[2][1]!=0):
        if(i[1][1] < 0):
            i[1][1] = i[1][1]*-1
            i[1][2] = i[1][2]*-1
            i[1][3] = i[1][3]*-1

        if(i[2][1] < 0):
            i[2][1] = i[2][1]*-1
            i[2][2] = i[2][2]*-1
            i[2][3] = i[2][3]*-1

        aux = i[1][1]
        #Cambiamos la primera funcion
        i[1][1] = i[1][1]*i[2][1]
        i[1][2] = i[1][2]*i[2][1]
        i[1][3] = i[1][3]*i[2][1]

        #Cambiamos la segunda funcion
        i[2][1] = i[2][1]*aux
        i[2][2] = i[2][2]*aux
        i[2][3] = i[2][3]*aux

        i[2][1] = i[1][1] - i[2][1]
        i[2][2] = i[1][2] - i[2][2]
        i[2][3] = i[1][3] - i[2][3]
    
    return i
   
eliminarX(i)
eliminarY(i)

z = i[2][3]/i[2][2]
y = (i[1][3]-i[1][2]*z)/i[1][1]
x = (i[0][3]-i[0][1]*y-i[0][2]*z)/i[0][0]

print("Resultado: ", "x - ", x, ". y - ", y, ". z - ", z)
```

#### Ejercicio 3[3 puntos]
El análisis del sentimiento de un texto consiste en asociar al texto un valor numérico que se obtiene de sumar/restar una cantidad por cada aparición de una palabra en el texto de acuerdo a un listado de palabras que tienen valores asociados. Más concretamente si se quiere obtener el análisis del sentimiento, se procesa el texto palabra a palabra:
  1) Se toma una palabra, y se busca en la lista de palabras que valor le corresponde.
  2) El valor se suma a un contador general.
  3) En caso de que la palabra no aparezca, se suma 0.
  4) Se pasa a la siguiente palabra
  5) Cuando ya no quedan más palabras por procesar del texto, se devuelve el número obtenido.

Se pide implementar un programa para calcular el análisis del sentimiento de un texto que será proporcionado a través de un fichero. El listado con las palabras y los valores asociados se proporcionará mediante otro fichero que deberá tener en cada línea del mismo una palabra y un valor separados ambos por un blanco y finalizados por salto de linea. Para realizar el procesamiento no se distinguirá entre mayúsculas y minúsculas. 

No se pueden usar ninguna función o método que realice directamente el procesamiento.

```python
valores=dict()   #Diccionario con valores


fichero = open("cuna.txt")   #abrimos fichero con texto cuna
fichero2=open("valores.txt")   #abrimos fichero2 con texto valores

cont=0  #ponemos variable cont a 0

for k in fichero2:   #recorremos fichero2 por lineas
    z=[]              #creamos lista
    z=k.split()       #rellenamos lista
    valores[z[0].lower()]=int(z[1])   #rellenamos diccionario con primera como clave y el segundo como valor
   
        
    



for i in fichero:    #recorremos fichero por lineas
    x=[]            #creamos lista
    x=i.split()       #rellenamos lista con palabras de la linea
    
    for j in x:    #por cada linea contamos cuantas veces aparece la palabra y buscamos su valor en el diccionario valores
        
        #Si alguna de las palabras está unida a un signo como coma, punto, dos puntos o punto y coma 
        #eliminamos este último valor
        if(j[len(j) - 1] == '.' or j[len(j) - 1] == ',' or j[len(j) - 1] == ';' or j[len(j) - 1] == ':'):
            j = j[0:len(j)-1]
            
        cont=cont +valores.get(j.lower(),0) 
    
       

print("El análisis del sentimiento da:")    
print(cont)   #mostramos contador
```
