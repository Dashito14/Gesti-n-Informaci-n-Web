Considerar el archivo __AccidentesBicicletas_2020.csv__. El archivo contiene la información sobre los accidentes de tráfico en la Ciudad de Madrid registrados por Policía Municipal, donde está implicada al menos una bicicleta, durante el año 2020.

Una descripción detallada de la estructura de los datos se puede encontrar en el siguiente enlace: https://datos.madrid.es/FWProjects/egob/Catalogo/Seguridad/Ficheros/Estructura_DS_Accidentes_Bicicletas_desde_2019.pdf

La información contenida por columnas es la siguiente:

|Campo|Significado|

Se pide crear funciones que realicen las siguientes operaciones:
* Leer los datos del archivo __AccidentesBicicletas_2020.csv__
* Consulta 1: Obtener y  mostrar por pantalla por cada Distrito (columna DISTRITO) y por cada tipo de persona(columna TIPO PERSONA), la suma total de accidentes que se han producido durante 2020, es decir:

|Distrito|Conductor| Peatón| Testigo|Viajero|
|----|-------------|-------------|-------------|-------------|
|Distrito1|Subtotal11 |--- |--- |Subtotal1n|  
|Distrito2|--- |--- |--- |--- | 
|   ...|--- |--- |--- |--- |     
|Distriton|Subtotal1n |--- |--- |Subtotalnn| 
     
* Consulta 2:Obtener y  mostrar por pantalla por cada Distrito (columna DISTRITO) cuántos accidentes acabaron en 07 asistencia sanitaria sólo en el lugar del accidente(Valor 07columna LESIVIDAD*) durante 2020, es decir:

|Distrito|Accidentes|

* Generar un documento __Salida.csv__ donde se almacene la información generada en la Consulta 1 de la siguiente forma:

    Distrito1, Conductor,Subtotal11, Peatón,Subtotal12,Testigo, Subtotal13,Viajero,Subtotal14
    
    Distriton, Conductor,Subtotaln1, Peatón,Subtotaln2,Testigo, Subtotaln3,Viajero,Subtotaln4
    
Tanto en lo que mostréis por pantalla como lo que escribáis en el archivo usad los nombres de los distritos que aparecen en los datos.
    

Para leer los archivos usa un código como el siguiente:

```python
def leerArchivo():
    import csv
    csvarchivo = open('AccidentesBicicletas_2020.csv',encoding="UTF8",errors='ignore')
    return csv.reader(csvarchivo, delimiter=";")

def consulta1(entrada):
    #Escribimos la cabecera y el numero de la consulta
    print('{:^63}'.format("CONSULTA 1"))
    print('{:^19}'.format("Distrito") + '{:^12}'.format("Conductor") + '{:^10}'.format("Peaton") + '{:^11}'.format("Testigo") + '{:^11}'.format("Viajero"))
    print("----------------------------------------------------------------")
    
    #Declaramos los diciionarios que utilizaremos para almacenar la cantidad de tipos de personas
    conductor=dict()
    peaton =dict()
    testigo=dict()
    viajero=dict()
    distrito=dict()
    
    #Por cada linea del archivo que hemos leido damos una iteracion
    for i in entrada:
        
        #Si es la primera linea de cabecera nos la saltamos
        if entrada.line_num == 1:
                continue # Saltar primera línea
        
        #Contamos cada tipo de persona "i[9]" (En el archivo a los viajeros se les llama pasajeros y a los peatones peatn)
        distrito[i[5]]=distrito.get(i[5],0)+1        
        if i[9]== "Conductor":
            conductor[i[5]]=conductor.get(i[5],0)+1
        elif i[9]=="Peatn":
            peaton[i[5]]= peaton.get(i[5],0)+1
        elif i[9]=="Testigo":
            testigo[i[5]]=testigo.get(i[5],0)+1
        elif i[9]=="Pasajero":
            viajero[i[5]]=viajero.get(i[5],0)+1
    
    #Por cada distrito almacenado iteramos
    for s in distrito:
        print ('{:<19}'.format(s) + '{:^12}'.format(str(conductor.get(s,0))) + '{:^10}'.format(str(peaton.get(s,0))) + '{:^11}'.format(str(testigo.get(s,0))) + '{:^11}'.format(str(viajero.get(s,0))))
        print("----------------------------------------------------------------")
    
    #Estas lineas simplemente son para la estetica de la salida ya que si no se juntan las salidas de todas las consultas
    print("----------------------------------------------------------------")
    print("----------------------------------------------------------------")
    print("----------------------------------------------------------------")
    print("----------------------------------------------------------------")
    print("----------------------------------------------------------------")
    print("----------------------------------------------------------------")
    
    
    
def consulta2(entrada):
    #Escribimos la cabecera y el numero de consulta de la salida
    print('{:^37}'.format("CONSULTA 2"))
    print('{:^19}'.format("Distrito") + '{:^18}'.format("Accidentes"))
    print("------------------------------------")
    
    #Declaramos el diccionario distrito
    distrito=dict()
    
    #Por cada linea del archivo de entrada damos una iteracion
    for i in entrada:
        #Si es la primera linea del archio (la cabecera) nos la saltamos
        if entrada.line_num == 1:
            continue # Saltar primera línea
        
        #Si la lesividad i[12] acaba en 07 sumamos uno al distrito correspondiente
        if(i[12]=="07"):
            distrito[i[5]]=distrito.get(i[5],0)+1  
    
    #Por cada distrito mostramos por pantalla su valor
    for x in distrito:
        print('{:<19}'.format(x) + '{:^18}'.format(str(distrito.get(x,0))))
        print("------------------------------------")
    
    #Estas lineas simplemente son para la estetica de la salida ya que si no se juntan las salidas de todas las consultas
    print("------------------------------------")
    print("------------------------------------")
    print("------------------------------------")
    print("------------------------------------")
    print("------------------------------------")
    print("------------------------------------")

    
    
    
def consulta3(entrada):
    #Cabecera con el numero de la consulta
    print('{:^24}'.format("CONSULTA 3"))
    
    #Repetimos el codigo de la consulta 1 puesto que no sabemos si se querria comprobar por separado
    #sin realizar las consultas en orden
    conductor=dict()
    peaton =dict()
    testigo=dict()
    viajero=dict()
    distrito=dict()
    for i in entrada:
        if entrada.line_num == 1:
                continue # Saltar primera línea

        distrito[i[5]]=distrito.get(i[5],0)+1        
        if i[9]== "Conductor":
            conductor[i[5]]=conductor.get(i[5],0)+1
        elif i[9]=="Peatn":
            peaton[i[5]]= peaton.get(i[5],0)+1
        elif i[9]=="Testigo":
            testigo[i[5]]=testigo.get(i[5],0)+1
        elif i[9]=="Pasajero":
            viajero[i[5]]=viajero.get(i[5],0)+1

    
    #Declaramos una lista vacia para almacenar los resultados
    data = []
     
    #Por cada distrito damos una iteracion
    for s in distrito:
        #Se almacena una lista por cada distrito
        data.append([s, "Conductor", conductor.get(s,0), "Peaton", peaton.get(s,0), "Testigo", testigo.get(s,0), "Viajero", viajero.get(s,0)] )
    
    #Escribimos en el archivo de salida
    with open('salida.csv', 'w') as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        writer.writerows(data)
        print("Fichero creado con exito")
        
        
entrada = leerArchivo()
consulta1(entrada)
entrada = leerArchivo()
consulta2(entrada)
entrada = leerArchivo()
consulta3(entrada)
```

#### Ejercicio 2[5 puntos]
Considerar el archivo __202006_movements.json__. El archivo contiene la información sobre los datos de los itinerarios de las bicicletas eléctricas del servicio BiciMAD 

Una descripción detallada de la estructura de los datos se puede encontrar en el siguiente enlace: https://opendata.emtmadrid.es/Documentos/Servicios-y-estructuras-Bicimad-V1-1.aspx

La información contenida por columnas es la siguiente:


* __id__: Identificador del movimiento.


* __user_day_code__: Código del usuario. Para una misma fecha, todos los movimientos de un mismo usuario, tendrán el mismo código, con el fin de poder realizar estudios estadísticos de las tendencias diarias de los usuarios.


* __idunplug_station__: Número de la estación de la que se desengancha la bicicleta. 

* __idunplug_base__: Número de la base de la que se desengancha la bicicleta. 

* __idplug_station__: Número de la estación en la que se engancha la bicicleta. 

* __idplug_base__: Número de la base en la que se engancha la bicicleta.

* __unplug_hourTime__: Franja horaria en la que se realiza el desenganche de la bicicleta. Por cuestiones de anonimato, se facilita la hora de inicio del movimiento, sin la información de minutos y segundos. Todos los movimientos iniciados durante la misma hora, tendrán el mismo dato de inicio.

* __travel_time__: Tiempo total en segundos, entre el desenganche y el enganche de la bicicleta.

* __track__: Detalle del trayecto realizado por la bicicleta entre la estación de partida y la de destino, en formato GeoJSON.

* __user_type__: Número que indica el tipo de usuario que ha realizado el movimiento. Sus posibles valores son:

    0: No se ha podido determinar el tipo de usuario 
    
    1: Usuario anual (poseedor de un pase anual)

    2: Usuario ocasional

    3: Trabajador de la empresa

* __ageRange__: Número que indica el rango de edad del usuario que ha realizado el movimiento. Sus posibles valores son:

    0: No se ha podido determinar el rango de edad del usuario 
    
    1: El usuario tiene entre 0 y 16 años

    2: El usuario tiene entre 17 y 18 años

    3: El usuario tiene entre 19 y 26 años

    4: El usuario tiene entre 27 y 40 años

    5: El usuario tiene entre 41 y 65 años 
    
    6: El usuario tiene 66 años o más

* __zip_code__: Texto que indica el código postal del usuario que ha realizado el movimiento.

Se pide crear funciones que realicen las siguientes operaciones:
* Leer los datos del archivo __202006_movements.json__
* Consulta 1: Obtener y  mostrar por pantalla por cada valor del campo __idunplug_station__, el número de bicicletas que fueron recogidas en dichos puntos de recogida, es decir:

|Punto de recogida|Bicicletas|
|----|-------------|
|Punto1|Total1|     
|PuntoN|TotalN|  
     
* Consulta 2:Obtener y  mostrar por pantalla por cada valor diferente del campo __ageRange__ , el número de bicicletas que fueron recogidas, es decir:

|Edad|Bicicletas|
|----|-------------|
|0|Total1|     
|1|Total2|
|---|---|
|6|Total6|

* Consulta 3: Generar una lista y mostrar una pantalla de aquellos puntos de recogida en los que ha habido usuarios que han recogido una bicicleta y los han usado también para dejarla posteriormente. Es decir, aquellos puntos donde existen registros que cumplen que __idunplug_station__ toma el mismo valor que __idplug_station__

Para leer los archivos usa un código como el siguiente:

```python
#Leemos el archivo
import json
leer = [json.loads(line) for line in open('202006_movements.json',encoding="utf8")]


def consulta1():
    #Numero de la consulta
    print("CONSULTA 1")
    
    #Declaramos un diccionario vacio
    dic=dict()
    
    #Por cada valor en el archivo JSON leido damos una iteracion
    for fila in leer:
        #Si la estacion ya estaba en el diccionario le sumamos uno
        if fila["idunplug_station"] in dic:
            dic[fila["idunplug_station"]]+=1
            
        #Si la estacion no estaba en el diccionario lo inicializamos a 1
        else:
            dic[fila["idunplug_station"]]=1
            
    #Ordenamos los valores
    dic = dict(sorted(dic.items()))
    
    #Mostramos por pantalla la cabecera
    print("Punto de recogida   Bicicletas") 
    print("------------------------------")
    
    #Por cada elemento en el diccionario mostramos por pantalla que elemento es y el numero de bicis recogidas
    for elemento in dic:
        print("\t",elemento,"\t\t",dic[elemento])
        
    
    #Estas lineas son solo esteticas para separar por pantalla las salidas de las consultas
    print("-------------------------------")
    print("-------------------------------")
    print("-------------------------------")
    print("-------------------------------")
    print("-------------------------------")
    print("-------------------------------")

    
    
def consulta2():
    print("CONSULTA 2")
    
    #Declaramos un diccionario vacio
    dic=dict()
    
    #Por cada fila en el archivo que hemos leido al prinicpio
    for fila in leer:
        #Si el rango de edad ya esta en el diccionario sumamos 1
        if fila["ageRange"] in dic:
            dic[fila["ageRange"]]+=1
            
        #Si no esta inicializamos el rango a 1
        else:
            dic[fila["ageRange"]]=1
            
    #Ordenamos el diccionario
    dic = dict(sorted(dic.items()))
    
    #Escribimos la cabecera
    print("\t Edad \t\t Bicicletas") 
    print("------------------------------------------")
    
    #Para cada elemento en el diccionario mostramos por pantalla el elemento y el numero de bicis
    for elemento in dic:
        print("\t",elemento,"\t\t",dic[elemento])
        
    #Estas lineas son solo para separar las consultas en la salida y que se vean mas facil
    print("------------------------------------------")
    print("------------------------------------------")
    print("------------------------------------------")
    print("------------------------------------------")
    print("------------------------------------------")
    print("------------------------------------------")
    
    
    

def consulta3(): 
    #Declaramos una lista vacioa
    lista = []
    
    #Por cada fila en el archivo leido anteriormente
    for fila in leer:
        
        #Si los valores de bicis sacadas y devueltas es el mismo metemos la estacion en la lista
        if fila["idunplug_station"] == fila["idplug_station"]:
            lista.append(fila["idunplug_station"])
            
    lista = set(lista)
    #Escribimos la cabecera
    print("CONSULTA 3")
    print("Estaciones con trayectos de origen y destino igual")
    print("--------------------------------------------------")
    
    #Por cada elemento en la lista la mostramos por pantalla
    for elem in lista:
        print("\t","\t","\t",elem)
        
        
        
#Se hacen todas las consultas en orden
consulta1()
consulta2()
consulta3()
```
